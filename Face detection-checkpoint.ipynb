```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
denis = cv2.imread('Denis_Mukwege.jpg',0)
solvay = cv2.imread('solvay_conference.jpg',0)
nadia = cv2.imread('Nadia_Murad.jpg',0)
```

```python
type(denis)
```

```python
plt.imshow(denis,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
        
    face_rects = face_cascade.detectMultiScale(face_img) #returns coordinates for the rectangle|
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eye(img):
    face_img = img.copy()
        
    eye_rects = eye_cascade.detectMultiScale(face_img) #returns coordinates for the rectangle|
    
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = detect_eye(denis)
plt.imshow(denis,cmap='gray')
```

```python
cap = cv2.VideoCapture(0)

while True : 
    ret,frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow('Video face detect',frame)
    
    k=cv2.waitKey(1)
    if k==27:
        break
        
cap.release()
cv2.destroyAllWindows()
                 
```
